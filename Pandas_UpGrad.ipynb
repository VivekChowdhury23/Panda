#### Hot chocolate is preferred in cold months. Identify the cold areas, so that the inventory team can keep the stock ready.

```python
weather['Year'] = pd.DatetimeIndex(weather['Date']).year
weather['Month'] = pd.DatetimeIndex(weather['Date']).month
weather['DayofMonth'] = pd.DatetimeIndex(weather['Date']).day

wea_mintemp = weather.groupby(by=["Month"]).mean()

wea_mintemp.sort_values("MinTemp").head()
```

#### Wind Chill Factor: (10*sqrt(velocity) - velocity + 10.5) * (33 - minTemp)

```python
from math import sqrt

def wci(x):
    velocity = x["WindGustSpeed"]
    minTemp = x["MinTemp"]
    return (10*sqrt(velocity) - velocity + 10.5)*(33 - minTemp)
```

```python
weather['WCI'] = weather.apply(wci, axis=1)

weather.head()
```

```python
weather_cold = weather.groupby(by="Month").mean()

weather_cold.sort_values("WCI", ascending=False).head()
```

### Pivot Tables

#### Using Pivot Tables find the avg monthly rainfall in the year 2016 of all the locations. This information can be used to predict the sales of tea in the year 2017.
